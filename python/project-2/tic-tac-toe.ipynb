```python
def display(grid):
    for i in range(3):
        print("")
        print(" __ __ __")
        print("\n|",end="")
        for j in range(3):
            print(grid[i][j],end=" |")
    print("\n __ __ __")
        
```

```python
def case1(grid,char):
    for i in range(3):
        empty = np.count_nonzero(grid[i]==' ')
        count = np.count_nonzero(grid[i]==char)
        if count==2 and empty!=0:
            index = np.where(grid[i]==' ')[0][0]
            row = i
            col = index
            return row,col
        else:
            empty = np.count_nonzero(grid[:,i]==' ')
            count = np.count_nonzero(grid[:,i]==char)
            if count==2 and empty!=0:
                index = np.where(grid[:,i]==' ')[0][0]
                row = index
                col = i
                return row,col
    empty = np.count_nonzero(grid.diagonal()==' ')
    count = np.count_nonzero(grid.diagonal()==char)
    if count==2 and empty!=0:
        index = np.where(grid.diagonal()==' ')[0][0]
        row = index
        col = i
        return row,col
    empty = np.count_nonzero(grid[:,::-1].diagonal()==' ')
    count = np.count_nonzero(grid[:,::-1].diagonal()==char)
    if count==2 and empty!=0:
        index = np.where(grid[:,::-1].diagonal()==' ')[0][0]
        row = index
        col = i
        return row,col
    return -1,-1
    
def case2(grid):
    l,k = [],0
    for i in range(3):
        for j in range(3):
            if grid[i][j]==' ':
                l.append(k)
            k+=1
    choice = random.choice(l)
    row = choice//3
    col = choice%3
    return row,col
```

```python
def play(grid,com_pts,hum_pts,com_turn,hum_turn):
    if com_pts==1:
        print("Computer Won!!!")
        return
    elif hum_pts==1:
        print("You Won!!!")
        return
    elif np.count_nonzero(grid)==9:
        print("It's a tie")
        return
    else:
        if hum_turn==1:
            print("It's your turn.Select a position from 0 to 8")
            position = int(input())
            row = position//3
            col = position%3
            while(grid[row][col]!=' '):
                print("Invalid position. Enter agian")
                position = int(input())
                row = position//3
                col = position%3
            grid[row][col]='X'
            if all([1 if grid[row][i]=='X' else 0 for i in range(3)]):
                hum_pts+=1
            if all([1 if grid[i][col]=='X' else 0 for i in range(3)]):
                hum_pts+=1
            if(position in [0,4,8] and all([1 if grid[i][i]=='X' else 0 for i in range(3)])):
                hum_pts+=1
            display(grid)
            play(grid,com_pts,hum_pts,1,0)
        else:
            row,col = -1,-1
            row,col = case1(grid,'X')
            if(row!=-1):
                grid[row][col]='O'
            else:
                row,col = case1(grid,'O')
                if(row!=-1):
                    grid[row][col]='O'
                    com_pts+=1
                else:
                    row,col = case2(grid)
                    grid[row][col]='O'          
            display(grid)
            play(grid,com_pts,hum_pts,0,1)
```

```python
grid = [[' ',' ',' '],[' ',' ',' '],[' ',' ',' ']]
grid=np.array(grid)
com_pts = 0
hum_pts = 0
hum_turn = 1
com_turn =0
play(grid,com_pts,hum_pts,com_turn,hum_turn)
```
