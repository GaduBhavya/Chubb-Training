```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('country_vaccination.csv')
ind_data = data[data['iso_code']=='IND']
usa_data = data[data['iso_code']=='USA']
```

```python
fig = plt.figure()
fig.set_figwidth(30)
fig.set_figheight(20)
plt.xticks(rotation='vertical')
plt.plot(usa_data['date'],usa_data['daily_vaccinations'],color='b',label="USA")
plt.plot(ind_data['date'],ind_data['daily_vaccinations'],color='g',label="INDIA")
plt.title("Daily Vaccinations in India and USA")
plt.legend(loc='upper center',prop={'size': 20})
plt.show()
```

```python
fig = plt.figure()
fig.set_figwidth(30)
fig.set_figheight(20)
plt.xticks(rotation='vertical')
plt.plot(usa_data['date'],usa_data['people_fully_vaccinated'],color='b',label="USA")
plt.plot(ind_data['date'],ind_data['people_fully_vaccinated'],color='g',label="INDIA")
plt.title("Complete Vaccinations in India and USA")
plt.legend(loc='upper center',prop={'size': 20})
plt.show()
```

```python
ind_jan_std = ind_data[pd.DatetimeIndex(ind_data['date']).month==1]['people_vaccinated'].std()
ind_feb_std = ind_data[pd.DatetimeIndex(ind_data['date']).month==2]['people_vaccinated'].std()
usa_jan_std = usa_data[pd.DatetimeIndex(usa_data['date']).month==1]['people_vaccinated'].std()
usa_feb_std = usa_data[pd.DatetimeIndex(usa_data['date']).month==2]['people_vaccinated'].std()
```

```python
months=['January','February']
fig, plt = plt.subplots()
index = np.arange(2)
bar_width = 0.25
plt.bar(index, [ind_jan_std,ind_feb_std], color = 'b', width = 0.25)
plt.bar(index+bar_width, [usa_jan_std,usa_feb_std], color = 'g', width = 0.25)
plt.set_xticks(index + bar_width / 2)
plt.set_xticklabels(months)
plt.legend(['INDIA','USA'])
```
