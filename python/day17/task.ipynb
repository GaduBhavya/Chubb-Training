```python
from matplotlib import colors
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def coefficient(x,y):
    no_of_obs = np.size(x)
    x_sum = np.sum(x)
    y_sum = np.sum(y)
    x_sq_sum = np.sum(x**2)
    xy_sum = np.sum(x*y)

    m = (no_of_obs*xy_sum - x_sum*y_sum)/(no_of_obs*x_sq_sum - x_sum*x_sum)
    c = (y_sum - m*x_sum)/no_of_obs
    return m,c

def regression(x,y,b,p):
    plt.scatter(x,y,color="red",marker="o",s=20)
    predicted_y = b[1] + b[0]*x
    plt.plot(x,predicted_y,color="g")
    if p==1:
        plt.xlabel("Open Data")
        plt.ylabel("Close Data")
    else:
        plt.xlabel("High Data")
        plt.ylabel("Close Data")
    plt.show()
```

```python
data = pd.read_csv("finance_data.csv")
open_data = np.array(data["Open"])
close_data = np.array(data["Close"])
high_data = np.array(data["High"])
```

```python
b = coefficient(open_data,close_data)
print("Standard Deviation(Open)=",np.std(open_data))
print("Standard Deviation(Close)=",np.std(close_data))
print("Regression coefficient=",b[0])
regression(open_data,close_data,b,1)
```

From the data it is observed that the standard deviation at open and close is 157.16 and 158.97 respectively. So it can be analyzed that the stock has become volatile at the time of closing. But there is no much exponential increase or decrease in the stock as data values are very close to regression line. 

And the value of R is 0.97 i.e., the opening and closing stocks are strongly interrelated. If the opening is high there is more chance that the stock closes profitably.

```python
b = coefficient(high_data,close_data)
print("Standard Deviation(High)=",np.std(high_data))
print("Standard Deviation(Close)=",np.std(close_data))
print("Regression coefficient=",b[0])
regression(high_data,close_data,b,0)
```

From the data it is observed that the R value is 0.99 i.e., it is almost equal to 1. So from this it is analyzed that the stocks are strongly interrelated. The closing stock is the highest stock of the day in most cases and in some cases the highest stock is more closer to the closing stock.
