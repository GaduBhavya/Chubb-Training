```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('case_time_series.csv')
```

```python
def graph(x,y):
    fig = plt.figure()
    fig.set_figwidth(30)
    fig.set_figheight(20)
    plt.scatter(x,y)
    poly = np.poly1d(np.polyfit(x,y,4))
    line = np.linspace(1,33113492,100)
    plt.plot(line,poly(line))
    plt.show()
    
def r_square(x,y):
    poly = np.poly1d(np.polyfit(x,y,4))
    return r2_score(y,poly(x))

def predict(x,y,value):
    poly = np.poly1d(np.polyfit(x,y,4))
    return poly(value)
```

```python
graph(data['Total Confirmed'],data['Total Deceased'])
print("R2 score is",r_square(data['Total Confirmed'],data['Total Deceased']))
print("For {} confirmed cases the predicted deceased cases are {}".format(100,predict(data['Total Confirmed'],data['Total Deceased'],100)))
```
